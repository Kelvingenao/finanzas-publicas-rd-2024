```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
gastos = pd.read_csv("gastos.csv")
ingresos = pd.read_csv("ingresos.csv")
```

```python


ingresos = ingresos.sort_values(by="Monto en USD", ascending=False)

# Estilo de gráfico más elegante
sns.set_style("whitegrid")  

# Crear la figura y los ejes
plt.figure(figsize=(10, 6))

# Gráfico de barras con color mejorado
ax = sns.barplot(data=ingresos, x="Categoría", y="Monto en USD", palette="coolwarm")

# Mostrar números completos en el eje Y (evita notación científica)
plt.ticklabel_format(style='plain', axis='y')

# Añadir valores en las barras
for i, v in enumerate(ingresos["Monto en USD"]):
    ax.text(i, v + 200000, f"{v/1_000_000:.1f}M", ha='center', fontsize=12, fontweight='bold', color='black')

# Personalización
plt.xlabel("Categoría", fontsize=14, fontweight='bold')
plt.ylabel("Monto en USD (Millones)", fontsize=14, fontweight='bold')
plt.title("Ingresos en USD por Categoría", fontsize=16, fontweight='bold')
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)

# Mostrar gráfico
plt.show()
```

```python
gastos = gastos.sort_values(by="Monto en USD", ascending=False)

# Estilo de gráfico más elegante
sns.set_style("whitegrid")  

# Crear la figura y los ejes
plt.figure(figsize=(10, 6))

# Gráfico de barras con color mejorado
ax = sns.barplot(data=gastos, x="Categoría", y="Monto en USD", palette="coolwarm")

# Mostrar números completos en el eje Y (evita notación científica)
plt.ticklabel_format(style='plain', axis='y')

# Añadir valores en las barras
for i, v in enumerate(gastos["Monto en USD"]):
    ax.text(i, v + 200000, f"{v/1_000_000:.1f}M", ha='center', fontsize=12, fontweight='bold', color='black')

# Personalización
plt.xlabel("Categoría", fontsize=14, fontweight='bold')
plt.ylabel("Monto en USD (Millones)", fontsize=14, fontweight='bold')
plt.title("Gastos en USD por Categoría", fontsize=16, fontweight='bold')
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)

# Mostrar gráfico
plt.show()
```

```python
# Datos en miles de millones (Billones)
monto_gasto = 16_040_000_000 / 1e9  # 16.04B
monto_ingreso = 49_580_000_000 / 1e9  # 49.58B

# Crear los datos para la gráfica (Ordenando de mayor a menor)
categorias = ["Ingreso", "Gasto"]
montos = [monto_ingreso, monto_gasto]  # Ahora Ingresos está primero

# Estilo de gráfico
sns.set_style("whitegrid")

# Crear la figura
plt.figure(figsize=(8, 5))

# Gráfico de barras
ax = sns.barplot(x=categorias, y=montos, palette=["#6B8E23", "#FF6F61"])  # Verde primero, rojo después

# Formatear el eje Y para que no use notación científica
ax.ticklabel_format(style='plain', axis='y')

# Añadir los valores en las barras con el formato en Billones (B)
for i, v in enumerate(montos):
    ax.text(i, v + 0.5, f"${v:.2f}B", ha='center', fontsize=12, fontweight='bold', color='black')

# Personalización del gráfico
plt.xlabel("")
plt.ylabel("Monto en USD (Billones)", fontsize=14, fontweight='bold')
plt.title("Comparación de Ingresos y Gastos", fontsize=16, fontweight='bold')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.figtext(0.5, -0.12, "Nota: Los montos están en miles de millones de dólares (B).", 
            ha="center", fontsize=11, color="#333333", fontweight='bold')
# Mostrar gráfico
plt.show()
```
